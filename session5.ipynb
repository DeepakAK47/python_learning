### Tuple is faster than  List

```python
import time  # you can use time function without importing this
```

```python
# WAP which shows tuple are faster than list
L = list(range(100000000))
L2 = tuple(range(100000000))

start = time.time()  # it gives the present time
for i in L:
    i*5
print("List take : ",time.time()-start)

# calculating time taken by the list
start2 = time.time()
for i in L2:
    i*10
print('The time takes by the tuple is : ',time.time()-start2)    
```

```python
# WAP to prove list takes more memory that tuple datatype
import sys
L = list(range(1000))
T = tuple(range(1000))
print('List take memory is : ',sys.getsizeof(L))
print('tuple takes the memory to store is : ',sys.getsizeof(T))
```

### Sets Operations

```python
# Union of both the set -> Note it will not make union with the duplicate elements
s1 = {1,2,3,4,5}
s2 = {5,6,7,8,9}
s1 | s2
```

```python
# Intersection 
s1 & s2
```

```python
# It is the set of all points in the s1 which are not present in the s2
s1 - s2
```

```python
# Symmetric difference(^)
s1 ^ s2  # IT will print all the elements except the common elements
```

```python
1 in s1  # Membership operation can also be applied here
```

```python
# len/min/max/sorted/sum -> All these operations are also available in the set 
sum(s1)
sum(s2)
```
