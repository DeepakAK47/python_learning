### Tuple is faster than  List

```python
import time  # you can use time function without importing this
```

```python
# WAP which shows tuple are faster than list
L = list(range(100000000))
L2 = tuple(range(100000000))

start = time.time()  # it gives the present time
for i in L:
    i*5
print("List take : ",time.time()-start)

# calculating time taken by the list
start2 = time.time()
for i in L2:
    i*10
print('The time takes by the tuple is : ',time.time()-start2)    
```

```python
# WAP to prove list takes more memory that tuple datatype
import sys
L = list(range(1000))
T = tuple(range(1000))
print('List take memory is : ',sys.getsizeof(L))
print('tuple takes the memory to store is : ',sys.getsizeof(T))
```

### {key : value for i in iterable conditions}

```python
# conversion of kilometers into 
distance = {'delhi' : 1000,'pune': 100,'hyderabad' : 3000}
distance
```

```python
# conversion of the killometers into miles 
{key:value*1.73 for (key,value) in distance.items()}  # it is very important multiplying both key value pairs
```

```python
# Their are two dictionaries one is storing the key and another is storing the values of that key merge them using the zip function
days = ['mon','tues','wed','thur','fri','sat','sun']
temp = [12,13,14,15,16,17,18]
# assigning their key values using zip function
{i:j for (i,j) in zip(days,temp)}
```
