```python
#(***)
# Use of constructor 
class student:
    def __init__(self,name,cgpa):  # this is called parameterized constructor
        self.name = name
        self.cgpa = cgpa
        print("Constructor is called")
    def get_cgpa(self):  # here self reference pointing to that object which call it. Ex ==> When it is called by std1 then here std1 is reference for this 
        return self.cgpa
std1 = student("Deepak",9.0)
std2 = student("Dilip",10.0)    

print(std1.name,std1.cgpa)
print(std2.name)

print(f"{std1.name} has {std1.get_cgpa()} cgpa") 
print(f"{std2.name} has {std2.get_cgpa()} cgpa")     
```

```python
#(***)
# WAP for printing the instance attributes
class Laptop:
    storage_type = "ssd"
    def __init__(self,RAM,Storage):
        self.RAM = RAM
        self.Storage = Storage
        
    def get_info(self):
            return print(f" laptop has Ram is {self.RAM} and its storage is {self.Storage} {self.storage_type}")
    @classmethod   # this is class decorator
    def get_storage_type(cls):
         print(f"storage type : {cls.storage_type}")

l1 = Laptop("256gb","512gb")
l2 = Laptop("2562gb","5122gb")
l3 = Laptop("2562gb","5122gb")

l1.get_info()
l2.get_info()
l3.get_info()

l1.get_storage_type()



# Note --> self can access instance attribute as well as class attributes.
# Note --> @classmethod can access only class attribute. They can not access instance attribute
```

```python
#WAP to calculate the discount on laptop using static method
class Laptop:
    @staticmethod
    def cal_discount(price,discount):
        final_price = price - ((price*discount)/100)
        print(f"The final price of laptop is {final_price}")

l1 = Laptop
l1.cal_discount(40000,10)
```

```python
# WAP for using concept of encapsulation
class Banking :
    def __init__(self,name,balance):
        self.name = name
        self.__balance = balance

    def get_bal(self):
        return self.__balance

    def set_val(self,newBal):
        self.__balance = newBal
        print(f"the new balance is : {self.__balance}")
        
b1 = Banking("Deepak",10000)
b2 = Banking("Vinay",12345678)
b3 = Banking("Vivek",1090283)
print(b1.get_bal())
b2.set_val(123)


# 👉 __balance uses name mangling to make the variable class-private, enforcing encapsulation, while balance is public and directly accessible.
```

```python
# WAP using the principle of inheritance
class Employee:
    start_time = "10am"
    end_time = "6pm"

class Teacher(Employee):
    def __init__(self,subject):
        self.subject = subject

t1 = Teacher("Maths")
print(t1.subject,t1.start_time,t1.end_time)        
```

```python
# WAP for use of super keyword
class Employee:
    start_time = "10am"
    end_time = "6pm"

class admin_staff(Employee):
    def __init__(self,role):
        self.role = role

class accountant(admin_staff):
    def __init__(self,salary,role):
        super().__init__(role)
        self.salary = salary
        self.role = role

acc1 = accountant(25_000,"CA")
print(acc1.role,acc1.start_time,acc1.end_time,acc1.salary)
```

```python
# WAP for multiple inheritance(*)
class Teacher:
    def __init__(self,salary):
        self.salary = salary
class Student:
    def __init__(self,name):
        self.name = name
class TA(Teacher,Student):
    def __init__(self,salary,name,cgpa):
        Teacher.__init__(self,salary)
        Student.__init__(self,name)
        self.cgpa = cgpa

ta1  = TA(25_000,"Deepak",8.6)                        

print(ta1.salary,ta1.name,ta1.cgpa)
```

```python
#(***)
# Doubt
# WAP for abstraction

from abc import ABC, abstractmethod  # We are importing abstraction module here

class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

class Lion(Animal):
    def make_sound(self):
        print("Roar!")

class Cow(Animal):
    def make_sound(self):
        print("mew")

lion = Lion()
lion.make_sound()

cow = Cow()
cow.make_sound()
```

```python
# WAP for function overriding

class Teacher():
    def get_ans(self):
        print("Deepak Singh is good in coding")

class Student(Teacher):
    def get_ans(self):
        print("Deepak singh is not good in coding")

s1 = Student()
s1.get_ans()
```

```python
# Duck typing in python
class Teacher():
    def get_ans(self):
        print("You are a good boy")

class Student():
    def get_ans(self):
        print("You are not a good boy")

t1 = Teacher()                
s1 = Student()

t1.get_ans()
s1.get_ans()
```
