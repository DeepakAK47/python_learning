```python
class Parent:

    def __init__(self,num):
        self.__num=num

    def get_num(self):
        return self.__num

class Child(Parent):

    def show(self):
        print("This is in child class")
        
son=Child(100)
print(son.get_num())
son.show()
```

```python
class Parent:

    def __init__(self,num):
        self.__num=num

    def get_num(self):
        return self.__num

class Child(Parent):

    def __init__(self,val,num):
        self.__val=val

    def get_val(self):
        return self.__val
        
son=Child(100,10)
print("Parent: Num:",son.get_num())
print("Child: Val:",son.get_val())
```

```python
# multilevel inheritance
class Student:
    def __init__(self,name1,roll_no,faculty_name):
        self.name1 = name1
        self.faculty_name = faculty_name
        self.roll_no = roll_no

    def name2(self):
        print('Student name is : ',self.name1)

    def roll_no2(self):
        print('Studnet roll number is : ',self.roll_no)

class Faculty(Student):
    def name3(self):
        print('Name of faculty is : ',self.faculty_name)

class Dean(Faculty):
    def Name(self):
        print('I am the dean of this college')
```

```python
d1 = Dean('Deepak',49,'Jyoti mam')
```

```python
print(d1.name3())
```

```python
# multiple inheritance
class Student:
    def __init__(self,name1,roll_no,faculty_name):
        self.name1 = name1
        self.faculty_name = faculty_name
        self.roll_no = roll_no

    def name2(self):
        print('Student name is : ',self.name1)

    def roll_no2(self):
        print('Studnet roll number is : ',self.roll_no)

class Faculty:
    def name3(self):
        print('Name of faculty is : ',self.faculty_name)

class Dean(Faculty,Student):
    def Name(self):
        print('I am the dean of this college')
```
