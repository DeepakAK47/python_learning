```python
# define aclass and use abstraction 
class Person:
    # constructor
    def __init__(self,name,gender,address):
        self.name = name
        self.gender = gender
        self.address = address

    
    def info(self):
        print('My name is : ',self.name)
        print('My gender is : ',self.gender)

    def address1(self):
        print(self.address._Address__city,self.address.pin,self.address.block)  # _Address -> Abstraction

class Address:
    def __init__(self,city,pin,block):
        self.__city = city  # city is private variable
        self.pin = pin
        self.block = block

    def get_city(self):
        return self.__city   # city is private variable

a1 = Address('Delhi',110094,'G')
p1 = Person('Deepak Singh','Male',a1)
```
