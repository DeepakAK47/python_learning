```python
# WAP for ATM machine using class
class Atm:
    # constructor
    def __init__(self):
        self.pin = ''
        self.bal = 0
    # menu function
    def menu(self):
        task = int(input('''
        Enter 1 for creating pin
        Enter 2 for changing pin
        Enter 3 for withdraw
        Enter 4 for balance
        Enter 5 for exit.
        '''))
    # condition for different cases
        if(task==1):
            self.create_pin()
        elif(task==2):
            self.change_pin()
        elif(task==3):
            self.withdraw()
        elif(task==4):
            self.balance()
        else:
            exit(self)
    # create_pin function
    def create_pin(self):
        user_pin = int(input('Enter the pin : '))
        self.pin  = user_pin
        user_bal = int(input('Enter the balance : '))
        self.bal = user_bal
        print('Pin and balance are generated successfully')
        self.menu()  # calling menu function again.
    def change_pin(self):
        old_pin = int(input('Enter the old pin : '))
        if(old_pin==self.pin):
            new_pin = int(input('Enter the new pin : '))
            self.pin = new_pin
            print('New pin is updated sucessufully')
        else:
            print('sale chor')
    def withdraw(self):
        old_pin = int(input('Enter the pin og your account : '))
        if(self.pin==old_pin):
            with_rupee = int(input('Enter the amount  :'))
            self.bal = self.bal-with_rupee;
            print('Transaction is completed sucessusfully')
        else: 
            print('Chor')
    def balance(self):
        curr_pin = int(input('Enter the pin'))
        if(self.pin==curr_pin):
            print('Your current balance is : ',self.bal)
        else:
            print('Enter incorrect pin')
        self.menu()
    def exit(self):
        pass

# Note -> create_pin, change_pin, withdraw all the methids. They are not function.        
```

```python
# __str__ magic method
class Fraction:
    def __init__(self,x,y):
        self.num = x
        self.den = y

    # __str__ magic method
    def __str__(self):
        return '{}/{}'.format(self.num,self.den)   # it defines what will print when we call print(obj)

    def __add__(self,other): # IT triger wehn we add - this between two objects
        new_num = self.num*other.den + self.den*other.num
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)

    def __sub__(self,other):
        new_num = self.num*other.den - self.den*other.num
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)

    #multiplication
    def __mul__(self,other):
        new_num = self.num*other.num
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)
```

```python
class Fraction:
    def __init__(self,x,y):
        self.num = x
        self.den = y

    def __str__(self):
        return '{}/{}'.format(self.num,self.den)

    def __add__(self,other):
        new_num = self.num*other.den + self.den*other.num
        new_den = self.den*other.den
        return Fraction(new_num, new_den)

    def __sub__(self,other):
        new_num = self.num*other.den - self.den*other.num
        new_den = self.den*other.den
        return Fraction(new_num, new_den)

    def __mul__(self,other):
        new_num = self.num*other.num
        new_den = self.den*other.den
        return Fraction(new_num, new_den)
```
